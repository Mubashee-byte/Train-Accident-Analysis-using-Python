# Casualties by Weather Condition

# Monthly Trends of Train Collisions
# 8.Group data by month to find the average number of collisions per month.


```python
import pandas as pd
df = pd.read_csv('train.csv')

se = df[df['Signal Error'] == 1].shape[0]                                         # Filter rows where Signal Error is 1

tc= df.shape[0]                                                                  # Count total collisions

sp = (se / tc) * 100                                                             # Calculate the percentage of signal error accidents

display('average number of collisions per month:',sp)                            # Print the final percentage

```

# Signal Error Analysis
# 9.Find out the percentage of collisions that occurred due to a signal error (Signal Error = 1).


```python
import pandas as pd    # Import Pandas Library

# Step 1: Load the dataset
df = pd.read_csv('train.csv')

# Step 2: Convert 'Day' column to DateTime format
dt = pd.to_datetime(df['Day'])

# Step 3: Extract Month from DateTime and create a new column 'Month'
df['Month'] = dt.dt.month

# Step 4: Group data by Month and count the number of collisions
mt = df.groupby('Month').size()

# Step 5: Print the total collisions month-wise
display(' total collisions month-wise',mt)

```

# 📌 **Matplotlib/Seaborn Tasks ( Data Visualization )**

# Yearly Collisions Trend
# 1. Create a line chart to show train collisions per year.


```python
import pandas as pd
import matplotlib.pyplot as plt   # For plotting graphs
df = pd.read_csv('train.csv')
# Yearly collisions data
collisions_per_year = df.groupby('Year').size()

# Plotting line chart
plt.plot(collisions_per_year.index, collisions_per_year.values, marker='o')
plt.title('Train Collisions per Year')
plt.xlabel('Year')
plt.ylabel('Number of Collisions')
plt.grid(True)
plt.show()
```

# Casualties by Cause
# 2. Use a bar chart (Column Chart) to visualize the number of casualties for each cause.


```python
import pandas as pd
import matplotlib.pyplot as plt                                                   # For plotting graphs
df = pd.read_csv("train.csv")
                                                                                  # Casualties by cause
cc = df.groupby('Cause')['Casualties'].sum()

                                                                                 # Plot bar chart
cc.plot(kind='bar', color='orange')
plt.title('Casualties by Cause of Collision')
plt.xlabel('Cause of Collision')
plt.ylabel('Number of Casualties')
plt.grid(True)
plt.show()
```

# Weather-wise Collision Distribution
# 3.Plot a donut chart showing the percentage of accidents under different weather conditions.


```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Load the dataset
df = pd.read_csv("train.csv")

# Step 2: Count accidents based on Weather condition
wc = df['Weather'].value_counts()

# Step 3: Plot the Donut Chart
plt.pie(wc, labels=wc.index, autopct='%1.1f%%', startangle=90)   # Plot Pie Chart with percentage & rotate 90 degree

# Step 4: Add Title
plt.title("Percentage of Accidents by Weather Condition", color='red')  # Add red color title

# Step 5: Remove x and y axis labels for a clean look
plt.xlabel(" ")
plt.ylabel(" ")

# Step 6: Display the chart
plt.show()

```

# Train Speed vs. Casualties
# 5. Create a column chart to compare casualties for different speed ranges (e.g., 0-50 km/hr, 51-100 km/hr, etc.).


```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("train.csv")

# Create Speed Ranges
speed_bins = [0, 50, 100, 150, 200]
speed_labels = ['0-50 km/hr', '51-100 km/hr', '101-150 km/hr', '151-200 km/hr']

# Add a new column "Speed_Range" in the dataset
df['Speed_Range'] = pd.cut(df['Train Speed'], bins=speed_bins, labels=speed_labels)

# Group by Speed Range and count total casualties
speed_data = df.groupby('Speed_Range')['Casualties'].sum()

# Plot Bar Chart
plt.bar(speed_data.index, speed_data.values, color='skyblue')
plt.title('Casualties in Different Speed Ranges')
plt.xlabel('Speed Range (in km/hr)')
plt.ylabel('Total Casualties')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Load Data
df = pd.read_csv("train.csv")

# Step 2: Create Speed Range and Count Casualties
data = df.groupby(pd.cut(df['Train Speed'], bins=[0, 50, 100, 150, 200], labels=['0-50', '51-100', '101-150', '151-200']))['Casualties'].sum()

# Step 3: Draw Bar Chart
plt.bar(data.index, data.values, color='orange')
plt.title('Casualties in Different Speed Ranges')
plt.xlabel('Speed Range (km/hr)')
plt.ylabel('Number of Casualties')
plt.show()
```
